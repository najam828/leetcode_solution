```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        result,new_result = [],[]
        for i in range(len(board)): 
            for j in range(len(board)):
                if board[i][j].isdigit() == True:
                    result.append(board[i][j])
                if board[j][i].isdigit() == True:
                    new_result.append(board[j][i])
            if len(result) != len(set(result)):
                return False
            if len(new_result) != len(set(new_result)):
                return False
            result.clear()
            new_result.clear()
        index = 0
        end = int(len(board)/3)
        result = []
        counter = 0
        for k in range(3):
            for i in range(len(board)):
                for j in range(index,end):
                    if board[i][j].isdigit() == True:
                        result.append(board[i][j])
                counter += 1
                if counter == 3:
                    if len(result) != len(set(result)):
                        return False
                    # print('True')
                    result.clear()
                    counter = 0     
            index += 3
            end += 3
        
        return True
```

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        elif len(nums) == 1:
            return 1
        new_num = list(set(nums))
        if  len(new_nums) == 1:
            return 1
        new_num.sort()
        ans = 0
        count = 0
        for i in range(len(new_num)):
            if i > 0 and new_num[i] == new_num[i-1]+1:
                count += 1
            else:
                count = 1
            ans = max(ans,count)
        return ans
                 
```

A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.

```python
Input = "A man, a plan, a canal: Panama"
new_input = "race a car"
new_new_input = " "
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = ''.join(e for e in s if e.isalnum())
        s = s.lower()
        return s == s[::-1]
```

```python
people_1 = [1,2]
limit_1 = 3
# Output: 1
# Explanation: 1 boat (1, 2)

people_2 = [3,2,2,1]
limit_2 = 3
# Output: 3
# Explanation: 3 boats (1, 2), (2) and (3)

people_3 = [3,5,3,4]
limit_3 = 5
# Output: 4
# Explanation: 4 boats (3), (3), (4), (5)

people_4 = [3,1,7]
limit_4 = 7
# Output = 2

people_5 = [5,1,4,2]
limit_5 = 6
# Output = 2
```

```python
class Solution:
    def numRescueBoats(self, people: List[int], limit: int) -> int:
        if (sum(people)== limit or sum(people)<limit) and len(people) <= 2:
            return 1
        if len(people) == 0:
            return 0
        people.sort()
        counter = 0
        initial = 0
        end = len(people)-1
        while (initial<=end):
            counter+=1
            if people[initial]+people[end] <= limit:
                initial+=1
            end-=1
            
        return counter 
        
```

```python
data = Solution().numRescueBoats(people_5,limit_5)
print(f"{data} answer")
```

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

```python
height_1 = [1,8,6,2,5,4,8,3,7]
Output_1 = 49
# Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. 
# In this case, the max area of water (blue section) the container can contain is 49.
# Example 2:

height_2 = [1,1]
Output_2 = 1

height_3 = [1,2]
Output_3 = 1

height_4 = [1,2,4,3]
Output_4 = 4

height_5 = [2,3,4,5,18,17,6]
Output_5 = 17
```

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        n = len(height)-1
        ini = 0
        Output = 0
        for _ in range(n):
            result = n-ini
            Output = max(min(height[ini],height[n])*result,Output)
            if height[ini] < height[n]:
                ini+=1
            else:
                n-=1
        return Output

```

The 1st place athlete's rank is "Gold Medal".
The 2nd place athlete's rank is "Silver Medal".
The 3rd place athlete's rank is "Bronze Medal".
For the 4th place to the nth place athlete, their rank is their placement number (i.e., the xth place athlete's rank is "x").
Return an array answer of size n where answer[i] is the rank of the ith athlete.

```python
# Example 1:

score_1 = [5,4,3,2,1]
Output_1 = ["Gold Medal","Silver Medal","Bronze Medal","4","5"]
# Explanation: The placements are [1st, 2nd, 3rd, 4th, 5th].
# Example 2:

score_2 = [10,3,8,9,4]
Output_2  = ["Gold Medal","5","Bronze Medal","Silver Medal","4"]
# Explanation: The placements are [1st, 5th, 3rd, 2nd, 4th].

score_3 = [1,2]
Output_3  = ["Silver Medal","Gold Medal"]
```

```python
class Solution:
    def findRelativeRanks(self, score: List[int]) -> List[str]:
        data = {0:"Gold Medal",1:"Silver Medal",2:"Bronze Medal"}
        n = len(score)
        result = [0] * n
        if n == 1 and score[0] > 1:
            return [data[0]]
        elif n == 2:
            max_value = max(score)
            min_value = min(score)
            result[score.index(max_value)] = data[0]
            result[score.index(min_value)] = data[1]
            return result
        elif n == 3:
            test = score.copy()
            max_value = max(score)
            min_value = min(score)
            test.remove(max_value),test.remove(min_value)
            result[score.index(max_value)] = data[0]
            result[score.index(test[0])] = data[1]
            result[score.index(min_value)] = data[2]
            return result

        test  = score.copy()
        for i in range(n):
            if not test:
                return result
            if len(test) == 1:
                if i in data:
                    result[score.index(test[0])]=data[i]
                else:
                    result[score.index(test[0])]=f'{n-i}'
                return result
            else:
                max_value = max(test)
                min_value = min(test)
                if i in data:
                    result[score.index(max_value)]=data[i]
                    result[score.index(min_value)]=f'{n-i}'
                else:
                    result[score.index(max_value)]=f'{i}'
                    result[score.index(min_value)]=f'{n-i}'
                test.remove(max_value)
                test.remove(min_value)
        return result

```

You are given an m x n binary matrix grid.

A move consists of choosing any row or column and toggling each value in that row or column (i.e., changing all 0's to 1's, and all 1's to 0's).

Every row of the matrix is interpreted as a binary number, and the score of the matrix is the sum of these numbers.

Return the highest possible score after making any number of moves (including zero moves).

```python
grid_1 = [[0,0,1,1],[1,0,1,0],[1,1,0,0]]
Output_1 =  39
# Explanation: 0b1111 + 0b1001 + 0b1111 = 15 + 9 + 15 = 39

grid_2 = [[0]]
Output_2 = 1

grid_3 = [[0,1],[0,0]]
Output_3 = 5

grid_4 = [[0,1],[1,1]]
Output_4 = 5
```

```python
class Solution:
    def matrixScore(self, grid: List[List[int]]) -> int:
        for row in grid:
            if row[0] == 0:
                for i in range(len(row)):
                    row[i] = 1 - row[i]
        def transpose(X: List[List[int]])->List[List[int]]:
            return  [list(row) for row in zip(*X)]
        data =  transpose(grid)
        result = []
        count_0,count_1 = 0,0
        for i in range(len(data)):
            for k in data[i]:
                if k == 0:
                    count_0+=1
                else:
                    count_1+=1
            if count_0 > count_1:
                result.append([0 if i==1 else 1 for i in data[i]])
            else:
                result.append(data[i])
            count_0,count_1=0,0
        new_data = transpose(result)
        result = 0
        for i in range(len(new_data)):
            test=''
            for j in new_data[i]:
                test+=str(j)
            result+=int(test,2)
        return result

```
